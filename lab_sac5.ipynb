```
f := x^4 - 4*x^3 + 7*x^2 - 5*x + 1
g := x^5 - x^4 - x^3 + x^2 + 5*x - 5
```

```
gcdStandart = PolynomialGCD[f,g]
```

```
{gcd, {u, v}} = PolynomialExtendedGCD[f, g]
Simplify[gcd == f * u + g * v]
```

```
ExtGCD[aa_, bb_] := Module[
    {a = aa, b = bb,
    u = 1, u1 = 0,
    v = 0, v1 = 1,
    quot, rem
    },
    
    While[Not[SameQ[b, 0]],
        quot = PolynomialQuotient[a, b, x];
        rem = PolynomialRemainder[a, b, x];
        {a, b} = {b, rem};
        {u, u1} = {u1, (u - u1*quot)//ExpandAll};
        {v, v1} = {v1, (v - v1*quot)//ExpandAll};

    ];
    {a, u, v} (*a - НОД *)
];
```

```
{extGCD, u, v} = ExtGCD[f, g];
"GCD =" extGCD
"u =" u
"v =" v
```

```
Simplify[extGCD == f * u + g * v]
```

Найдем старший коэффициент многочлена extGCD

```
coefficient := CoefficientList[extGCD[[1]], x]//Last
```

Нормируем полученные многочлены

```
gcdNormalize = (extGCD / coefficient)//ExpandAll
```

```
uNormalize = (u / coefficient)//ExpandAll
```

```
vNormalize = (v / coefficient)//ExpandAll
```

Сравним результаты полученной функции и стандартной функции

```
SameQ[gcdStandart, -1 * gcdNormalize]
```

Разложение Безу

```
bezu = (uNormalize * f + vNormalize * g)//ExpandAll
```

```
SameQ[bezu, gcdNormalize]
```
